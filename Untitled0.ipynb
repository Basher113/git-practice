```python
reshaped_df = df.pivot(index="DATE", columns="TAG", values="POST")
reshaped_df
```

```python
reshaped_df.fillna(0, inplace=True)
```

```python
reshaped_df.isna().values.any()
```

```python
import matplotlib.pyplot as plt
roll_df = reshaped_df.rolling(window=6).mean()
plt.figure(figsize=(18,12))
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(0, 35000)
plt.xlabel("Date entry of the language in stackoverflow", fontsize=16)

for column in reshaped_df.columns:
  plt.plot(reshaped_df.index, roll_df[column], linewidth=3, label=reshaped_df[column].name)
plt.legend(fontsize=25)
```
